```python
## Renaminng columns
df=df.rename(columns={'Date':'Sales_date'})
df.head()
```

```python
import pandas as pd
url ='https://www.cricbuzz.com/cricket-team/india/2/stats'
df = pd.read_html(url)
df[0].to_excel('BAtting.xlsx')
```

```python
df[0].describe()
```

```python
import matplotlib.pyplot as plt
```

```python
cricket=df[0]
cricket_runs =cricket.groupby('PLAYER')['AVG'].sum()
cricket_runs.plot(
    kind='bar',
    figsize=(10, 6),
    xlabel='Players',
    ylabel='AVG',
    title='Players vs Runs',
    color='blue'
)

plt.xticks(rotation=45, ha='right')  # Rotate player names for better visibility
plt.tight_layout()  # Adjust layout to avoid label overlap
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group data by player and sum runs
cricket = df[0]  # Assuming df[0] contains the PLAYER and RUNS columns
cricket_runs = cricket.groupby('PLAYER')['AVG'].sum()

# Sort players by runs for a cleaner chart
cricket_runs = cricket_runs.sort_values(ascending=False)

# Create the plot
plt.figure(figsize=(12, 7))
sns.lineplot(
    x=cricket_runs.index,
    y=cricket_runs.values,
    palette="viridis"  # Use a beautiful color palette
)

# Customize the chart
plt.title('Players vs Batting Average (AVG)', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Players', fontsize=14, labelpad=10, color='darkgreen')
plt.ylabel('Batting Average (AVG)', fontsize=14, labelpad=10, color='darkgreen')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Remove the top and right spines for a cleaner look
sns.despine()

# Add average values above bars for clarity
for i, avg in enumerate():
    plt.text(i, avg + 1, f"{avg:.2f}", ha='center', va='bottom', fontsize=10, color='black')
# Show the plot
plt.tight_layout()
plt.show()
```
