```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
df = pd.DataFrame(iris['data'])
```

```python
df.columns = iris.feature_names
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(df)
```

```python
## Apply pca
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
pca_X_scaled=pca.fit_transform(X_scaled)
```

```python
plt.scatter(pca_X_scaled[:,0],pca_X_scaled[:,1],c=iris.target)
```

```python
## Agglomerative cluserting
## to contruct a dendogram
import scipy.cluster.hierarchy as sc
## plot dendogram
plt.figure(figsize=(20,7))
plt.title("Dendograms")

sc.dendrogram(sc.linkage(pca_X_scaled ,method='ward'))
plt.title('dendogram')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2,linkage='ward')
cluster.fit(pca_X_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_X_scaled[:,0],pca_X_scaled[:,1], c=cluster.labels_)
```
